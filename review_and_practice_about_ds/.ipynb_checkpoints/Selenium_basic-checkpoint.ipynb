```python
from selenium import webdriver
```

```python
#open the browser
driver = webdriver.Chrome()
```

```python
#move the browser
driver.get('http://www.naver.com')
```

```python
#change the size or window
driver.set_window_size(800,600)
```

```python
#move the scroll by using js code
script = 'window.scrollTo(200,300);'
driver.execute_script(script)
```

```python
# save current window in object
main_window = driver.current_window_handle
main_window
```

```python
# open the new window tab
script = 'window.open("https://www.google.com");'
driver.execute_script(script)#not focus in this tab
driver.get('http://daum.net/')#main tab which is on focus
```

```python
#check the opened window
windows = driver.window_handles
windows
```

```python
# change focus of tab
driver.switch_to_window(windows[1])
driver.get('http://facebook.com')#google -> facebook
driver.switch_to_window(main_window)#focus back to daum
```

```python
#refreshing the page
script = 'location.reload();'
driver.execute_script(script)
```

```python
#take alert
script = 'alert("selenium test");'
driver.execute_script(script)
```

```python
# alert function into object
alert = driver.switch_to_alert()
alert.text
```

```python
# accept the alert
alert.accept()
```

```python
#confirm is similar with alert but it also has cancel option.
script = 'confirm("confirm?");'
driver.execute_script(script)
```

```python
#input data
driver.find_element_by_css_selector('#q').send_keys('hello world!')
```

```python
#click the button
driver.find_element_by_css_selector('.ico_pctop.btn_search').click()
```

```python
alert.accept()
alert.dismiss()
driver.close()
driver.quit()
```

practice selenium in gmarket best seller

```python
from selenium import webdriver
```

```python
driver = webdriver.Chrome()
driver.get('http://corners.gmarket.co.kr/Bestsellers')
```

```python
bestlist_selector = driver.find_elements_by_css_selector('#gBestWrap.best-list')[1]
items= bestlist_selector.find_elements_by_css_selector('li')[:10]
for item in items:
    idx = item.find_element_by_css_selector("p").text
    content = item.find_element_by_css_selector(".itemname").text
    o_price = int(item.find_element_by_css_selector(".item_price .o-price").text[:-1].replace(",",""))
    s_price = item.find_element_by_css_selector(".item_price .s-price").text
    s_price = int(s_price.split(" ")[0][:-1].replace(",",""))
    print("{}\n제목 - {}\n정상가 - {}\n할인가 - {}\n".format(idx, content, o_price, s_price))
```

```python
import time
from selenium import webdriver
from PIL import Image as pil
```

```python
driver = webdriver.Chrome()
driver.get('https://www.youtube.com/watch?v=tXOJIBPwSyo&t=200s')
time.sleep(1)
driver.save_screenshot('screenshot.png')
```

```python
pil_im = pil.open('screenshot.png')
print(pil_im.size)
pil_im
```

```python
pil_im.thumbnail((512,512))
pil_im.save('screenshot_thumbnail.png')
pil.open('screenshot_thumbnail.png')
```

```python
element = driver.find_element_by_css_selector('#movie_player')
location = element.location
size = element.size
location,size
```

```python
pil_im = pil.open('screenshot.png')

left = location['x'] 

top = location['y']

right = left + size['width'] 

bottom = top + size['height'] 

area = (left, top, right, bottom)

area
```

```python
pil_im = pil_im.crop(area)
```

```python
pil_im.save('screenshot2.png')
pil.open('screenshot2.png')
```
