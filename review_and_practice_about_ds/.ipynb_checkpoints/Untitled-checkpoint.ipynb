```python
import scipy as sp
```

```python
def make_regression2(n_samples,bias,noise,random_state):
    np.random.seed(random_state)
    X = np.random.rand(n_samples).reshape(n_samples,1)
    w = np.random.randint(100)+ np.random.rand()
    mu = 0
    std = noise
    rv = sp.stats.norm(mu,std)
    y = bias + w*X + rv.rvs(n_samples).reshape(n_samples,1)
    return X,y,w
```

```python
X,y,w = make_regression2(100,3,2,1)
```

```python
plt.scatter(X, y,s=100)
plt.grid(True)
plt.show()
```

- 상관관계란 x의 값이 달라지면, y의 분포가 달라지는 것을 의미합니다.
- 인과관계와는 다릅니다.
