#### 샘플 왜도(skewness)
- 평균과의 거리를 세제곱을 이용하여 구한 특징값을 의미
- 왜도가 0미연 대칭이므로 왜도는 분포의 대칭성을 알아보는 지표

#### 샘플 첨도(kurtosis)
- 평균과의 거리를 네제곱을 이용하여 구한 특징값을 의미
- 첨도는 데이터가 중앙에 몰려있는 정도를 정밀하게 비교하기 위해 쓰인다.

#### 샘플 모멘트(moment)
- 분산 왜도 첨도를 구하기 위해 제고브 세제고브네제곱을 하는 것처럼 더 고차의 제곱을 이용하여 구한 특징값을 moment라고 한다. N제곱을 이용하여 구한 모멘트를 N차 모멘트라고 한다. 따라서 평균은 1차 모멘트 분산은 2차 모멘트 왜도는 3차 모멘트 첨도는 4차 모멘드이다.

1~4차 모멘트가 같으면 같은 확률적 데이터로 보는 것이 일반적이다.

##### 샘플 평균도 확률 변수이므로 기댓값이 존재한다. 샘플 평균의 기댓값은 원래 확률 변수의 기댓값과 일치한다.

```python
x=np.random.randint(100,size=100)
np.mean(x)
y1=np.mean(x[:20])
y2=np.mean(x[20:40])
y3=np.mean(x[40:60])
y4=np.mean(x[60:80])
y5=np.mean(x[80:])
arr=[y1,y2,y3,y4,y5]
ls=[i for i in arr]
np.mean(ls)#샘플 평균의 기댓값
np.mean(x)#원래의 확률 변수의 기댓값
np.mean(ls)==np.mean(x)
```

##### 샘플 평균의 분산은 원래 확률 변수의 분산과 다음과 같은 관계를 가진다.
