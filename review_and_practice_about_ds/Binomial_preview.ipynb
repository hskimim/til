### 이항 분포
- 성공확률이 theta인 베르누이 시도를 N번 하는 경우
- N times of Bernoulli trial

```python
N=10
theta=0.6
rv=sp.stats.binom(N,theta)
rv
```

```python
xx=np.arange(N+1)
plt.bar(xx,rv.pmf(xx),align='center')
plt.ylabel('P(x)')
plt.title('pmf of binomial distribution')
plt.show()
```

```python
np.random.seed(0)
x=rv.rvs(100)
x
```

```python
sns.countplot(x)
plt.title('simulation of binomial distribution')
plt.xlabel('sample')
plt.show()
```

```python
y=np.bincount(x,minlength=N+1)/float(len(x))
df=pd.DataFrame({'theory': rv.pmf(xx),'simulation': y}).stack()
df=df.reset_index()
df.columns=['sample','type','ratio']
df.pivot('sample','type','ratio')
df[:5]
```

```python
sns.barplot(x='sample',y='ratio',hue='type',data=df)
plt.show()
```
