### Scikit-Learn 의 샘플 데이터
- load 계열 명령
    - 패키지에 처음부터 저장되어 다운로드 없이 바로 쓸 수 있는 것이다.
- fetch 계열 명령
    - 인터넷에서 다운로드 받아서 홈 디렉토리 아래의 서브 디렉토리에 저장한 후 추후 불러들이는 데이터
- make 계열 명령
    - 원하는 특성을 가진 가상의 데이터가 필요할 수 있는데 이때 이러한 가상 데이터를 생성한다.
- 데이터 세트의 형식
    - data 
        - input arguement form of matrix
    - target
        - outpus argument form of vector
    - feature_names
        - data's name list
    - target_names
        - target's name list
    - DESCR
        - kind of docstring about data(option)
