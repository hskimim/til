```python
from selenium import webdriver
import time
```

```python
def navertitle1(pageNum=10):
    driver = webdriver.Chrome()
    header = []
    for i in range(1,pageNum+1):
        url = 'http://news.naver.com/main/list.nhn?mode=LSD&mid=sec&listType=title&sid1=105&date=20180627&page={}'.format(i)
        driver.get(url)
        time.sleep(2)
        titles1 = driver.find_elements_by_css_selector('.type02 li')
        for title1 in titles1:
            header.append(title1.find_element_by_css_selector('a').text)
    driver.quit()
    return header
```

```python
def navertitle2(pageNum=10):
    driver = webdriver.Chrome()
    header = []
    for i in range(1,pageNum+1):
        url = 'http://news.naver.com/main/list.nhn?mode=LSD&mid=sec&sid1=105&date=20180627&page={}'.format(i)
        driver.get(url)
        titles1 = driver.find_elements_by_css_selector('.type06_headline li')
        titles2 = driver.find_elements_by_css_selector('.type06 li')
        for title1 in titles1:
            header.append(title1.find_element_by_css_selector('dt:not(.photo) > a').text)
        for title2 in titles2:
            header.append(title2.find_element_by_css_selector('dt:not(.photo) > a').text)
    driver.quit()
    return header
```
