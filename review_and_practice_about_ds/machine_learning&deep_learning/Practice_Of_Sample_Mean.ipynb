```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sample=np.random.randn(100)
ls=[np.mean(np.random.randn(100)) for i in range(50)]
```

```python
# plt.xlim(400,600)
# plt.xlim(0,1000)
plt.hist(sample,bins=10)
print('sample mean: {}'.format(np.mean(sample)))
plt.hist(ls,bins=10)
print('sample mean of sample mean: {}'.format(np.mean(ls)))
plt.show()
```

```python
import scipy as sp

rv = sp.stats.norm()
type(rv)
```

```python
rv = sp.stats.norm(loc=1, scale=0.3)
```

```python
import matplotlib.pylab as plt
xx = np.linspace(-8, 8, 10000)
pdf = rv.pdf(xx)
plt.plot(xx, pdf)
plt.show()
```

```python
xx = np.linspace(-8, 8, 100)
cdf = rv.cdf(xx)
plt.plot(xx, cdf)
plt.show()
```

```python
import seaborn as sns
rv = sp.stats.norm(loc=1, scale=0.3)
sns.distplot(rv.rvs(size=10000, random_state=0))
plt.xlim(-8, 8)
plt.show()
```

```python
rv = sp.stats.norm(loc=1, scale=0.3)
sample = rv.rvs(size=(10000),random_state=0)
sample_mean = np.mean(sample)
sample_SD=np.std(sample,ddof=1)
sample_mean,sample_SD
```
