##### Pivot_table
데이터 열 중에서 두 개를 key로 사용하여 데이터를 선택하는 방법을 말한다.
행 인텍스와 열 인덱스는 하나의 데이터를 찾는 키 역할을 한다. 즉 이값으로 데이터가 유일하게 결정되어야 한다.

#### groupby
특정 조건에 맞는 데이터가 하나 이상, 즉 그룹을 이루는 경우에는 그룹 분석을 해야 한다.

```python
import seaborn as sns
iris=sns.load_dataset('iris')
iris.tail()
```

```python
def peak_to_peak_ratio(x):
    return x.max()/x.min()
```

```python
iris.groupby(iris.species).agg(peak_to_peak_ratio)
```

```python
iris.groupby(iris.species).describe().T
```

```python
def max3(df):
    return df.sort_values(by='petal_length')[:3]
```

```python
iris.groupby(iris.species).apply(max3)
```

pivot이란 데이터 열중에서 두 개를 키로 사용하여 데이터를 선택하는 방법을 말한다.
index, column, value-- index, column이 두 개의 key를 의미

그룹 분석이란 특정 조건에 맞는 데이터가 하나 이상 즉, 그룹을 이루는 경우에는 그룹 분석을 해야 한다. 2개의 그룹을 잡고 연산을 실행(분석)

pivot명령과 groupby 명령의 중간 성격을 가지는 pivot_table명령도 있다. pivot_table명령은 groupby 명령처럼 그룹 분석을 하지만 최종적으로는 pivot 명령처럼 피봇테이블을 만든다. pivot처럼 index,column을 잡고 table을 만들지만 aggfunc를 통해서 그룹분석(연산)도 실행한다.
