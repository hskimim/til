```python
from scipy.cluster.hierarchy import linkage, dendrogram
np.set_printoptions(precision=5, suppress=True) 

np.random.seed(4711)  # for repeatability of this tutorial
a = np.random.multivariate_normal([10, 0], [[3, 1], [1, 4]], size=[100,])
b = np.random.multivariate_normal([0, 20], [[3, 1], [1, 4]], size=[50,])
X = np.concatenate((a, b),)
print(X.shape)

plt.scatter(X[:,0], X[:,1], s=50)
plt.show()
```

```python
Z = linkage(X, 'ward')
```

```python
Z[:20]
# 앞에 있는 두 개의 컬럼은 순서이다. 예로 들어서 52,53번 데이터가 하나의 클러스터를 
# 구성했다는 뜻이다. 그다음은 거리이다.0.0451은 52,53번 데이터 사이의 거리, 마지막은
# 원소의 갯수이다. 
# 몇 번 데이터는 클러스터 넘버를 의미한다. 즉, 1개짜리 클러스터를 넘는 순간, 151번 
# 클러스터가 된다. 따라서 넘버 부분을 보게 되면, 150을 넘어가는 것도 있고, 마지막
# 원소의 갯수 파트를 보게 되도, 2를 넘는 것이 존재하게 된다.
```
