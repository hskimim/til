```python
thata1=0.5
rv1=sp.stats.bernoulli(thata1)
```

```python
thata2=0.9
rv2=sp.stats.bernoulli(thata2)
```

```python
xx=[0,1]
plt.subplot(221)
plt.bar(xx,rv1.pmf(xx))
plt.xlim(-1, 2)
plt.ylim(0, 1)
plt.subplot(222)
plt.bar(xx,rv2.pmf(xx))
plt.xlim(-1, 2)
plt.ylim(0, 1)
plt.xticks([0, 1],["x=0", "x=1"])
plt.xlabel("sample value")
plt.title("pmf of Bernoulli distribution")
plt.show()
```

```python
x1=rv1.rvs(10,random_state=0)
x2=rv2.rvs(100,random_state=0)
```

```python
plt.subplot(221)
sns.countplot(x1)
plt.title('simulation of x1')
plt.xlabel('sample')
plt.subplot(222)
sns.countplot(x2)
plt.title('simulation of x2')
plt.xlabel('sample')
plt.show()
```

#### 베르누이 분포의 모수추정(parameter estimation)
베르누이 분포에서 모수는 theta값이다.
theta의 approximate value는 N1(1이 나온 횟수)/N(전체 데이터의 수)이다.
위의 예시에서 보이듯, N이 클 수록, 즉 sample data가 많아지면 많아질 수록, theory's parameter와 가까워진다.

#### 베르누이 분포의 활용
- 1) 분류 예측 문제의 출력 데이터가 두 개의 값으로 구분되는 카테고리 값인 경우.
- 2) 입력 데이터가 0 or 1, True or False인 경우
