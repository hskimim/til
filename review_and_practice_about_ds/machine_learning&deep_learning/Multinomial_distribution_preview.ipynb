### Multinomial distribution
- Before, we've talked that multiple trial summation of Bernoulli trial forms binomial distribution. And now, we're gonna talk about multiple trial outcome of categorical distribution. that is `Multinomial distribution.`

- parameter : theta = (theta_1,...,theta_k)
- probability variable : x = (x_1,...,x_k)
- Mu(x;N;theta) = `combination(N,x)*(theta_k)**(x_k) = combination(N,x1,...,x_k)*(theta_k)**(x_k)`
- moment : 
    - first moment: E[x_k] = N*theta_k
    - second moment: Var[x_k] = N*theta_k(1-theta_k)
