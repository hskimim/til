```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
data=np.random.randint(10,size=10)
plt.plot(data)
plt.show()
```

```python
xtick=range(0,100,10)
plt.plot(xtick,data)
plt.show()
```

```python
plt.plot(data,'r*:')
plt.show()
```

```python
plt.xlim(-5,20)
plt.ylim(-10,30)
plt.plot(data)
plt.show()
```

```python
data=np.random.randint(3,size=5)
plt.plot(data)
plt.xticks(range(5))
plt.yticks([1,2])
plt.show()
```

```python
data=np.random.randint(10,size=10)
plt.grid(True)
plt.plot(data)
plt.show()
```

```python
data1 = np.random.randint(5, size=5)
data2 = np.random.randint(5, size=5)
data3 = np.random.randint(5, size=5)
xticks = range(0,5)
plt.plot(xticks,data1,'ro-',xticks,data2,'g^:',xticks,data3,'b*--')
plt.show()
```

```python
data1=np.random.randint(5,size=5)
data2=np.random.randint(5,size=5)
plt.plot(data1,label='data1')
plt.plot(data2,label='data2')
plt.legend(loc=0)
plt.show()
```

```python
data=np.random.randint(5,size=5)
plt.plot(data,label='data')
plt.legend(loc=0)
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('title')
plt.show()
```

```python
data=np.random.randint(5,size=5)
plt.figure(figsize=(4,3))
plt.plot(data)
plt.show()
```

```python
data1=np.random.randint(5,size=5)
data2=np.random.randint(5,size=5)
plt.subplot(221)
plt.plot(data1)
plt.subplot(222)
plt.plot(data2)
plt.show()
```

```python
y=np.random.randint(10,size=5)
x=range(5)
plt.bar(x,y)
plt.show()
```

```python
y=np.random.randint(5,10,size=5)
x=range(5)
error=np.random.randint(1,5,size=5)
plt.bar(x,y,yerr=error,align='center',alpha=0.2)
plt.show()
```

```python
# uniform (균등분포)
x1 = np.random.rand(500)
y1 = np.random.rand(500)
plt.subplot(331)
plt.scatter(x1, y1)
# gaussian (정규분포)
x2 = np.random.randn(500)
y2 = np.random.randn(500)
plt.subplot(332)
plt.scatter(x2, y2)
# randint (uniform)
x3 = np.random.randint(1,100,size=500)
y3 = np.random.randint(1,100,size=500)
plt.subplot(333)
plt.scatter(x3, y3)
plt.show()
```

```python
# 50개의 구간을로 잘라 해당되는 데이터의 수를 카운트한 후에 y축의 값을 normalized 한다
x1 = np.random.rand(1000)
plt.subplot(211)
plt.hist(x1, bins=50)
x2 = np.random.randn(1000)
plt.subplot(212)
plt.hist(x2, bins=50)
plt.show()
```
